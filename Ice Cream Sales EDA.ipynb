```python
df=pd.read_csv('Monthly Ice Cream Shop Sales by Flavor.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'],df['Total Profit'], marker='o', color='blue', label='Total Profit')
plt.title('Total Profit Across All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Month'],df['Total Units'])
plt.title('Total Units Sold Per Month')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.show()
```
